```python
class data_cleaning:
    def __init__(self):
        pass
    def message_cleaning(self, message):
        
            Test_punc_removed = [char for char in message if char not in string.punctuation] # removing punctuations
            
            Test_punc_removed_join = ''.join(Test_punc_removed)
            # removing stop words
            Test_punc_removed_join_clean = [word.lower().strip() for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
            final_join = ' '.join(Test_punc_removed_join_clean)
            return final_join
    
        
    def apply_to_column(self, data_column_text):
        data_processed = data_column_text.apply(self.message_cleaning)
        return data_processed
```

```python
data_clean_obj = data_cleaning()
df_cnn['article_cnn'] = data_clean_obj.apply_to_column(df_cnn['article'])
df_fox['article_fox'] = data_clean_obj.apply_to_column(df_fox['article'])
```

```python
# merge the two files on date column.
df_final = pd.merge(left=df_cnn, right=df_fox, how='inner',left_on='date', right_on='date')
```

```python
df_final.head()
```

# **TF-IDF**

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer()
def get_cosine_similarity_TF_IDF(article):
    tfidf_matrix = tfidf_vectorizer.fit_transform(article)
    cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
    return cosine_sim[0][1]
```

```python
all_articles = list(zip(df_final['article_cnn'], df_final['article_fox']))
```

```python
cosine_similarity_list = [get_cosine_similarity_TF_IDF(article) for article in all_articles]
```

```python
df_final['cosine_similarity_TF_IDF'] = cosine_similarity_list
```

```python
df_final = df_final.sort_values(by='cosine_similarity_TF_IDF',ascending=False)
df_final['cosine_similarity_TF_IDF'] = df_final['cosine_similarity_TF_IDF']*100
df_final['cosine_similarity_TF_IDF'] = round(df_final['cosine_similarity_TF_IDF'],2)
```

```python
df_final[['date', 'article_x', 'article_y',
       'cosine_similarity_TF_IDF']].head()
```

# **word2vec**

```python
import gensim.downloader
glove_vectors = gensim.downloader.load('word2vec-google-news-300')
```

```python
csim = []
for i in df_final.index:
  sent1 = df_final.iloc[i]['article_cnn']
  sent2 = df_final.iloc[i]['article_fox']

  sen_1_words = [w for w in sent1.split() if w in glove_vectors.vocab]
  sen_2_words = [w for w in sent2.split() if w in glove_vectors.vocab]
  sim = glove_vectors.n_similarity(sen_1_words, sen_2_words)
  csim.append(sim)
```

```python
df_final['cosine_similarity_WORD2VEC'] = csim
```

```python
df_final['cosine_similarity_WORD2VEC'] = df_final['cosine_similarity_WORD2VEC']*100
df_final['cosine_similarity_WORD2VEC'] = round(df_final['cosine_similarity_WORD2VEC'],3)
```

```python
df_final = df_final.sort_values(by=['cosine_similarity_TF_IDF','cosine_similarity_WORD2VEC'], ascending=[False, True]).reset_index(drop=True)
```

```python
df = df_final[['date', 'article_x', 'article_y',
       'cosine_similarity_TF_IDF', 'cosine_similarity_WORD2VEC']]
df.rename(columns={'article_x':'CNN', 'article_y':'FOX'}, inplace=True)
```
